```python
import random
import statistics as s
#Q1 Write a program to count the number of vowels and consonants present in an input string


def count_vowels_consonants(n):
 vowels=['a','e','i','o','u']# vowels list
 v=0
 c=0
 for i in range(len(n)):
   for j in range(5):
    if n[i]==vowels[j-1]:#checking if it is an vowel
      v=v+1
      break# breaks the loop if found
   else:
    c=c+1
 return v,c
n= input("enter a string")
v,c=count_vowels_consonants(n)
print("number of vowels",v)
print("number of consonants",c)


#Q2 Write a program that accepts two matrices A and B as input and returns their product AB
#Check if A and B are multipliable; if not, return error message


def mul_matrix(rows_a,cols_a,rows_b,cols_b,A,B):
  if cols_a!=rows_b:# checks if multiplacation is possible if not returns an error
    return "error"
  AB=[]
  for i in range(rows_a):# creates a row in which the multiplied row is stored
        row = []
        for j in range(cols_b):# now creates to 2 more loops to the the AB element
            num = 0
            for k in range(cols_a):
                num += A[i][k] * B[k][j]
            row.append(num)# appends all the num in row
        AB.append(row) # appends all the row in AB
  return AB


rows_a = int(input("Enter number of rows for Matrix A: ")) #asking the no of rows and columns
cols_a = int(input("Enter number of columns for Matrix A: "))

print("Enter elements of Matrix A:")
A = [] # Write a sequence of elements in a row in one line
for i in range(rows_a):
    row = list(map(int, input().split()))# takes the sequence and splits it and makes it int and create a list
    A.append(row) # and append the list in A creating a matrix


rows_b = int(input("Enter number of rows for Matrix B: "))
cols_b = int(input("Enter number of columns for Matrix B: "))

print("Enter elements of Matrix B:")
B = []
for i in range(rows_b):
    row = list(map(int, input().split()))
    B.append(row)

AB= mul_matrix(rows_a,cols_a,rows_b,cols_b,A,B)

if AB == "error":#checks if output is present
    print("Matrix multiplication not possible")
else:
    print("Resultant Matrix:")# displays the output
    for row in AB:
        print(row)



#Q3 Write a program to find the number of common elements bw two lists
#The lists contain integers

def common_elements(list1, list2):
  common=[]
  for i in range(len(list1)):# taking 2 for loops to travse all the elements and check if they have a common element
    for j in range(len(list2)):
      if list1[i]==list2[j]:
        common.append(list1[i])#appends the element in common
  return common
print("Enter the first list")#takes the list input
X=list(set(list(map(int,input().split()))))
print("Enter the second list")
Y=list(set(list(map(int,input().split()))))
#if u want unique common element convert xy to set and back to list
com=common_elements(X,Y)
print(len(com))


#Q4 Write a program that accepts a matrix as input and returns its transpose

def trans(rows_a,cols_a,A):
  T = [] # create a transpose matrix
  for j in range(cols_a):
      row = [] # creates a list to store row
      for i in range(rows_a):
          row.append(A[i][j]) # see that j and i are interchanges in a transpose matrix
      T.append(row) # appens the row to transpose matrix
  return T



rows_a = int(input("Enter number of rows for Matrix A: "))# asks the no of rows and columns
cols_a = int(input("Enter number of columns for Matrix A: "))

print("Enter elements of Matrix A:") # matrix as input code
A = []
for i in range(rows_a):
    row = list(map(int, input().split()))
    A.append(row)
transpose= trans(rows_a,cols_a,A)
print("Resultant Matrix:") # display transpose
for row in transpose:
    print(row)

#Q5 Generate a list of 100 random numbers bw 100 and 150.
#Find the mean, meadian and mode for these numbers

def mmm():
    random.seed(42)# so we get the same random variable everytime
    a = []
    for i in range(100):
        num = random.randint(100, 150)# generates random numbers
        a.append(num)# appends in list a
    mean = s.mean(a) # finds mean meedian and mode
    median = s.median(a)
    mode = s.mode(a)
    return mean, median, mode
mean,median,mode=mmm()
print("mean",mean)# displays it
print("median",median)
print("mode",mode)




```
